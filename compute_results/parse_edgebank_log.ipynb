```python
import numpy as np
import pandas as pd
import argparse
from argparse import Namespace
```

```python
log_file = '../EdgeBank/log/second-baseline-2.log'
```

```python
with open(log_file, 'r') as f_log:
    log_list = f_log.readlines()
log_list = [log.strip() for log in log_list]
log_list
```

```python
method_list= []
auc_list = []
ap_list = []

for log in log_list:
    if 'Namespace' in log:
        args = eval(log)
        method_list.append('{}-{}-{}'.format(args.mem_mode, args.data, args.neg_sample))
        auc_list.append([])
        ap_list.append([])
    if 'Old nodes -- auc_inherent' in log:
        auc_list[-1].append(eval(log.split(':')[-1]))
    if 'Old nodes -- ap_inherent' in log:
        ap_list[-1].append(eval(log.split(':')[-1]))   
```

```python
for i in range(len(method_list)):
    print(method_list[i])
    print('auc: {:5.3} | {}'.format(np.mean(auc_list[i]), np.std(auc_list[i])))
    print('ap:  {:5.3} | {}'.format(np.mean(ap_list[i]), np.std(ap_list[i])))
```
